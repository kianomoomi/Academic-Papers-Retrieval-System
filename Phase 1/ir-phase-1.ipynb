```python
# 6 points


def clean_data(text: str):
    """Preprocesses the text with tokenization, case folding, stemming and lemmatization, and punctuations

    Parameters
    ----------
    text : str
        The title or abstract of an article

    Returns
    -------
    list
        A list of tokens
    """

    # TODO: tokenize, case_folding, stem, lemmatize, punctuations
    text_casefold = text.casefold()
    text_alphabetic = ''.join([char for char in text_casefold if char.isalpha() or char == ' '])
    tokens = word_tokenize(text_alphabetic)
    ps = PorterStemmer()
    for i, token in enumerate(tokens):
        tokens[i] = ps.stem(token)
    lemmatizer = WordNetLemmatizer()
    for i, token in enumerate(tokens):
        tokens[i] = lemmatizer.lemmatize(token)
    return tokens
# return ["an", "abstract", "is", "a", "summary", "of", "the", "main", "article"]
clean_data("An abstract is a summary of the main article.")
```

```python
def encode_gammacode(num):
    if num == 0:
        return 'x'
    binary = bin(num)[2:]
    offset = binary[1:]
    unary = '1' * len(offset) + '0'
    return unary + offset

def decode_gammacode(gamma):
    if gamma == 'x':
        return 0
    unary = gamma.split('0')[0]
    offset = gamma[len(unary) + 1:]
    return int('1' + offset, 2)
```

```python
def encode_variable_byte(number):
    bytes_list = []
    while True:
        bytes_list.insert(0, number % 128)
        if number < 128:
            break
        number //= 128
    bytes_list[-1] += 128
    return ''.join('{0:08b}'.format(byte) for byte in bytes_list)

def decode_variable_byte(variable_byte):
    bytes_list = []
    while True:
        bytes_list.append(int(variable_byte[:8], 2))
        variable_byte = variable_byte[8:]
        if bytes_list[-1] >= 128:
            bytes_list[-1] -= 128
            break
    result = 0
    for byte in bytes_list:
        result *= 128
        result += byte
    return result
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.positions = []


class PositionalIndex:
    def __init__(self):
        self.root = TrieNode()

    def add(self, token, doc_id, position, is_title):
        node = self.root
        for char in token:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.positions.append((doc_id, position, is_title))

    def remove(self, doc_id):
        def remove_helper(node, doc_id):
            if not node:
                return
            for child in node.children.values():
                remove_helper(child, doc_id)
            node.positions = [pos for pos in node.positions if pos[0] != doc_id]
        remove_helper(self.root, doc_id)

    def to_dict(self):
        result = {}
        def to_dict_helper(node, token):
            if not node:
                return
            if node.positions:
                result[token] = node.positions
            for char, child in node.children.items():
                to_dict_helper(child, token + char)
        to_dict_helper(self.root, '')
        return result

    @classmethod
    def from_dict(cls, data):
        index = cls()
        for token, postings in data.items():
            for doc_id, position, is_title in postings:
                index.add(token, doc_id, position, is_title)
        return index

    def to_dict_gamma_code(self):
        result = {}
        def to_dict_gamma_code_helper(node, token):
            if node.positions:
                positions = [(new_doc_ids[doc_id], position, is_title) for doc_id, position, is_title in node.positions]
                encoded_positions = []
                last_doc_id = 0
                for doc_id, position, is_title in positions:
                    encoded_doc_id = encode_gammacode(doc_id - last_doc_id)
                    encoded_position = encode_gammacode(position)
                    encoded_positions.append((encoded_doc_id, encoded_position, '1' if is_title else '0'))
                    last_doc_id = doc_id
                result[token] = encoded_positions
            for char, child in node.children.items():
                to_dict_gamma_code_helper(child, token + char)
        to_dict_gamma_code_helper(self.root, '')
        return result

    @classmethod
    def from_dict_gamma_code(cls, data):
        index = cls()
        for token, postings in data.items():
            last_doc_id = 0
            for encoded_doc_id, encoded_position, is_title in postings:
                doc_id = decode_gammacode(encoded_doc_id) + last_doc_id
                position = decode_gammacode(encoded_position)
                last_doc_id = doc_id
                for key, value in new_doc_ids.items():
                    if value == doc_id:
                        doc_id = key
                        break
                index.add(token, doc_id, position, is_title == '1')
        return index

    def to_dict_variable_byte(self):
        result = {}
        def to_dict_variable_byte_helper(node, token):
            if node.positions:
                positions = [(new_doc_ids[doc_id], position, is_title) for doc_id, position, is_title in node.positions]
                encoded_positions = []
                last_doc_id = 0
                for doc_id, position, is_title in positions:
                    encoded_doc_id = encode_variable_byte(doc_id - last_doc_id)
                    encoded_position = encode_variable_byte(position)
                    encoded_positions.append((encoded_doc_id, encoded_position, '1' if is_title else '0'))
                    last_doc_id = doc_id
                result[token] = encoded_positions
            for char, child in node.children.items():
                to_dict_variable_byte_helper(child, token + char)
        to_dict_variable_byte_helper(self.root, '')
        return result

    @classmethod
    def from_dict_variable_byte(cls, data):
        index = cls()
        for token, postings in data.items():
            last_doc_id = 0
            for encoded_doc_id, encoded_position, is_title in postings:
                doc_id = decode_variable_byte(encoded_doc_id) + last_doc_id
                position = decode_variable_byte(encoded_position)
                last_doc_id = doc_id
                for key, value in new_doc_ids.items():
                    if value == doc_id:
                        doc_id = key
                        break
                index.add(token, doc_id, position, is_title == '1')
        return index
```

```python
# 12 points

def construct_positional_indexes(corpus):
    """
    Get processed data and insert words in that into a trie and construct postional_index and posting lists after wards.

    Parameters
    ----------
    corpus: str
        processed data 

    Return
    ----------
    docs: 
        list of docs with specificied id, title, abstract.
    """
    index = PositionalIndex()
    docs = []
    all_titles_clean = corpus[:len(df)]
    all_abstracts_clean = corpus[len(df):]
    for i in range(len(df)):
        title = all_titles_clean[i]
        abstract = all_abstracts_clean[i]
        doc = {}
        doc['paperId'] = df['paperId'][i]
        doc['title'] = title
        doc['abstract'] = abstract
        docs.append(doc)
        for j in range(len(title)):
            index.add(title[j], doc['paperId'], j, True)
        for j in range(len(abstract)):
            index.add(abstract[j], doc['paperId'], j, False)
    return index, docs


index, docs = construct_positional_indexes(all_text)
```

```python
def calculate_docs_average_length():
    docs_title_average_length = 0
    docs_abstract_average_length = 0
    for doc in docs:
        docs_title_average_length += len(doc['title'])
        docs_abstract_average_length += len(doc['abstract'])
    docs_title_average_length /= len(docs)
    docs_abstract_average_length /= len(docs)
    return docs_title_average_length, docs_abstract_average_length

    

def okapi_bm25(title_query, abstract_query, max_result_count, weight, print_snippet, k1=1.2, b=0.75):
    title_scores = {}
    abstract_scores = {}
    docs_title_average_length, docs_abstract_average_length = calculate_docs_average_length()
    for query_word in title_query.split():
        title_query = title_query.replace(query_word, correct_text(query_word))
    title_query_tokens = clean_data(title_query)
    for token in title_query_tokens:
        posting_list = get_posting_list(token)
        for doc_id, positions in posting_list.items():
            if positions['title'] != []:
                if doc_id not in title_scores:
                    title_scores[doc_id] = 0
                tf = len(positions['title'])
                idf = math.log(len(docs) / len(posting_list), 10)
                for doc in docs:
                    if doc['paperId'] == doc_id:
                        current_doc = doc
                        break
                document_score = idf * tf * (k1 + 1) / (tf + k1 * (1 - b + b * len(current_doc['title']) / docs_title_average_length))
                title_scores[doc_id] += document_score


    for query_word in abstract_query.split():
        abstract_query = abstract_query.replace(query_word, correct_text(query_word))
    abstract_query_tokens = clean_data(abstract_query)
    for token in abstract_query_tokens:
        posting_list = get_posting_list(token)
        for doc_id, positions in posting_list.items():
            if positions['abstract'] != []:
                if doc_id not in abstract_scores:
                    abstract_scores[doc_id] = 0
                tf = len(positions['abstract'])
                idf = math.log(len(docs) / len(posting_list), 10)
                for doc in docs:
                    if doc['paperId'] == doc_id:
                        current_doc = doc
                        break
                document_score = idf * tf * (k1 + 1) / (tf + k1 * (1 - b + b * len(current_doc['abstract']) / docs_abstract_average_length))
                abstract_scores[doc_id] += document_score
    scores = {}
    for doc_id in set(title_scores.keys()).union(set(abstract_scores.keys())):
        title_score = title_scores.get(doc_id, 0)
        abstract_score = abstract_scores.get(doc_id, 0)
        final_score = weight * abstract_score + (1 - weight) * title_score
        scores[doc_id] = final_score
    scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    if max_result_count == -1:
        max_result_count = len(scores)
    top_results = scores[:max_result_count]
    if print_snippet:
        for doc in top_results:
            print("id: ", doc[0])
            print("Title: ", df[df['paperId'] == doc[0]]['title'].values[0][:50])
            print("Abstract: ", df[df['paperId'] == doc[0]]['abstract'].values[0][:50])
            print("*******************")
    return top_results
```

```python
from typing import List


def calculate_precision(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the precision of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The precision of the predicted results
    """
    precision = 0.0

    # TODO: Calculate precision here
    for i in range(len(actual)):
        precision += len(set(actual[i]) & set(predicted[i])) / len(predicted[i])
    precision /= len(actual)

    return precision
```

```python
def calculate_recall(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the recall of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The recall of the predicted results
    """
    recall = 0.0

    # TODO: Calculate recall here
    for i in range(len(actual)):
        recall += len(set(actual[i]) & set(predicted[i])) / len(actual[i])
    recall /= len(actual)
    

    return recall
```

```python
def calculate_F1(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the F1 score of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The F1 score of the predicted results    
    """
    f1 = 0.0

    # TODO: Calculate F1 here
    precision = calculate_precision(actual, predicted)
    recall = calculate_recall(actual, predicted)
    f1 = 2 * precision * recall / (precision + recall)

    return f1
```

```python
def calculate_MAP(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Mean Average Precision of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The Mean Average Precision of the predicted results
    """
    map = 0.0

    # TODO: Calculate MAP here
    for i in range(len(actual)):
        sum = 0
        for j in range(len(predicted[i])):
            if predicted[i][j] in actual[i]:
                sum += len(set(actual[i]) & set(predicted[i][:j+1])) / (j+1)
        map += sum / len(actual[i])
    map /= len(actual)
    

    return map
```

```python
def cacluate_NDCG(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Normalized Discounted Cumulative Gain (NDCG) of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The NDCG of the predicted results
    """
    ndcg = 0.0

    # TODO: Calculate NDCG here
    for i in range(len(actual)):
        dcg = 0
        for j in range(len(predicted[i])):
            if predicted[i][j] in actual[i]:
                if j == 0:
                    dcg += 1
                else:
                    dcg += 1 / np.log2(j+1)
        ideal_dcg = 0
        for j in range(len(actual[i])):
            if j == 0:
                ideal_dcg += 1
            else:
                ideal_dcg += 1 / np.log2(j+1)
        ndcg += dcg / ideal_dcg

    ndcg /= len(actual)
    

    return ndcg
```

```python
def cacluate_MRR(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Mean Reciprocal Rank of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The MRR of the predicted results
    """
    MRR = 0.0

    # TODO: Calculate MRR here
    for i in range(len(actual)):
        sum = 0
        for j in range(len(predicted[i])):
            if predicted[i][j] in actual[i]:
                sum += 1 / (j+1)
                break
        MRR += sum
    MRR /= len(actual)

    return MRR
```
